```python
# 08. Categorical Classification
df = pd.read_csv('laptops.csv')
df.head(5)
```

```python
brand_nation = {
    'Dell' : 'U.S.',
    'Apple' : 'U.S.',
    'Acer' : 'Taiwan',
    'HP' : 'U.S.',
    'Lenovo' : 'China',
    'Alienware' : 'U.S.',
    'Microsoft' : 'U.S.',
    'Asus' : 'Taiwan'    
}
```

```python
df['brand'].map(brand_nation)
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
df.head(5)
```

```python
# 11. Groupby
# analyze by category
nation_groups = df.groupby('brand_nation')
type(nation_groups) # NEW datatype!
```

```python
nation_groups.count()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
nation_groups.last()
# the first data of the each data (also the last)
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```
