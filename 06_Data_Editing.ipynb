```python
toeic = pd.read_csv('toeic.csv')
toeic
```

```python
toeic['합격 여부'] = False
part = (toeic['LC'] > 250) & (toeic['RC'] > 250)
sums = toeic['LC'] + toeic['RC'] > 600
```

```python
toeic.loc[part & sums, '합격 여부'] = True
```

```python
import pandas as pd

df = pd.read_csv('toeic.csv')

pass_total = df['LC'] + df['RC'] > 600
pass_both = (df['LC'] >= 250) & (df['RC'] >= 250)
df['합격 여부'] = pass_total & pass_both

# 정답 출력
df
```
