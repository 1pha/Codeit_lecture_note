```python
import pandas as pd
import numpy as np
```

```python
iphone_df = pd.read_csv('iphone.csv')
iphone_df
```

```python
iphone_df = pd.read_csv('iphone.csv', index_col=0)
iphone_df
```

```python
iphone_df['iPhone 8', '메모리']
# unavailable, does not work as a list
```

```python
# "loc" necessary!, when one row line is important, one data-point is important
# df.loc[record/index , feature]
iphone_df.loc['iPhone 8', '메모리']
type(iphone_df.loc['iPhone 8', '메모리'])
```

```python
# one row = 'Series'
iphone_df.loc['iPhone 8', :]
iphone_df.loc['iPhone 8'] # loc : row first, column conditional
type(iphone_df.loc['iPhone 8', :])

# multiple rows = DataFrame
# df.loc[ [rows] ], [rows] ex. ['iPhone 8', 'iPhone X'], ['iPhone 8':'iPhone X']
iphone_df.loc[['iPhone 8', 'iPhone X']]
type(iphone_df.loc[['iPhone 8', 'iPhone X']])
```

```python
# one column = 'Series' : one line = 'Series'
iphone_df['출시일'] # calling one column doesn't need loc
type(iphone_df['출시일'])


# multi columns need list of columns -> df[ [column list] ]
iphone_df[['출시일', '메모리']] # row first, so multiple columns need list of cols
type(iphone_df[['출시일', '메모리']]) # multilines = DataFrame

iphone_df['출시일':'메모리'] # doesn't work absolutely
iphone_df.loc[:, '출시일':'메모리'] # loc necessarily needs row num
type(iphone_df.loc[:, '출시일':'메모리'])


# always remember we need 'row > column'
# one row
# multi row : 1. by list 2. by indexing > they work the same

# one column
# multi column : 1. by list      works same as one column
#                2. by indexing  need "loc"!
```

```python
# 13. Conditional Indexing
iphone_df['디스플레이'] > 5
type(iphone_df['디스플레이'] > 5)
# still, series
```

```python
# turn lights on : row
iphone_df[[True, False, True, False, True, True, False]] # format used frequently ... alias
iphone_df.loc[[True, False, True, False, True, True, False]]
iphone_df.loc[[True, False, True, False, True, True, False], :] # the above actually means

# if value suffers, rest of them sets to false
#iphone_df[[True, False, True, False, True]] > this doesn't work
iphone_df.loc[[True, False, True, False, True]]
```

```python
# turn on lights on columns
iphone_df.loc[:, [True, False, False, True]]
```

```python
iphone_df['디스플레이'] > 5 # pd.Series of boolean > we can use this as condition

iphone_df.loc[iphone_df['디스플레이'] > 5]
iphone_df.loc[iphone_df['디스플레이'] > 5, [True, False]]
```

```python
iphone_df['Face ID'] == "Yes" # condition for each row

iphone_df.loc[iphone_df['Face ID'] == "Yes"]
iphone_df.loc[iphone_df['Face ID'] == "Yes", [True, True, False]]
```

```python
# using multiple condition
condition1 = iphone_df['Face ID'] == "No"
condition2 = iphone_df['디스플레이'] > 5

iphone_df.loc[condition1 & condition2]
iphone_df.loc[condition1 | condition2]
iphone_df[condition1]
```

```python
# 18. Index Location : iloc(using number)
iphone_df.iloc[2, 4] # 2nd row, 4th column
```

```python
iphone_df.iloc[[1, 3], [1, 4]] # row : [1st, 3rd], column : [1st, 4th]
iphone_df.iloc[3:, 1:4]
iphone_df.iloc[3:, [1, 4]]
```
